## Datasets
### Dataset Introducation
Our analysis based on two separate datasets:  
[US Colleges and Universities](https://public-us.opendatasoft.com/explore/dataset/us-colleges-and-universities/information)  
This dataset is published by HIFLD and consists of all Post Secondary Education facilities acknowledged by American official departments. It covers 7686 American universities and colleges located at 50 states, as well as Puerto Rico and other assorted U.S. territories. We leverages this dataset to acquire universities’ fundamental information including location, population, name, etc. 

[Computer Science Professors from 55 Top US Schools](https://drafty.cs.brown.edu/professors/#!data)  
This dataset published by Prof. Huang in Brown University. It contains information of over 2,000 computer science professors at top universitiesm, which include the names, institution, degrees obtained, and when they joined the university, for professors in the traditional role that involves both research and teaching. 

###  Dataset Preprocessing
####  1. University dataset
(a) We eliminated all universities whose population is less than 100 because we consider those universities as not active enough to provide enough information for our subsequent analysis.   
(b) We eliminated all universities whose total enrollment is less than 0 for the same reason.  
(c) We eliminated all universities which are located outside of the 50 American states because we want to mainly concentrate on the commonsense USA--Continental United States. Because of the variance of development, policy and some other external factors, it’s meaningless to compare the information of those universities which sit alone on islands in Pacific Ocean with that of universities sit in the 50 states.  
(d) We only kept columns 'NAME', 'STATE', 'POPULATION', 'COUNTY', 'NAICS_DESC', 'LEVEL_', 'TOT_ENROLL' which are essential attributes for our proposed questions. Additionally, we split column 'Geo Point' into 'LATITUDE' and 'LONGITUDE' and saved them for convenience. We eliminated all tuples which have 'Not Available' or 'NULL' in aforementioned fields.  
(e) Finally, we got 5,897 universities for following task.  
#### 2. Professor dataset
This dataset has been cleaned so we only removed some unnecessary columns, such as 'PhotoURL', 'Last Updated By'.

The following pie charts shows the distribution of univerisites and popluations among states. 

```python
college_state_dis = {}
college_state_popluation_dis = {}
for i in range(df_college.shape[0]):
    state = df_college.loc[i,'STATE']
    college_state_dis[state] = college_state_dis.get(state, 0) + 1
    college_state_popluation_dis[state] = college_state_popluation_dis.get(state, 0) + df_college.loc[i,'POPULATION']
college_state_labels = []
college_state_size = []
college_state_population_size = []
for k, v in  college_state_dis.items():
    college_state_labels.append(k)
    college_state_size.append(v)
    college_state_population_size.append(college_state_popluation_dis[k])

fig1, (ax1,ax2) = plt.subplots(nrows = 2)
ax1.pie(college_state_size, labels=college_state_labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax1.set_title('Distribution of univerisites among states')

ax2.pie(college_state_population_size, labels=college_state_labels, autopct='%1.1f%%',
        shadow=False, startangle=90)
ax2.set_title('Distribution of univerisity student popluation among states')
fig1.set_size_inches(15.0,30.0)

plt.show()
```

As we can see, California has most univerites (10.5%) and student popluation (12.8%). And top 5 states owns over one fourth of the total popluation, which shows the unequal distribution of universites in each state.

### 2. Professor dataset
Import the cleaned dataset.

```python
professor_data_url = 'https://raw.githubusercontent.com/ShengruiLYU/Professor-Data/master/Computer%20Science%20Professors%20from%2055%20Top%20US%20Schools%20(Brown%20University%20HCI%20project)%20-%20Data.csv'
df_prof = pd.read_csv(professor_data_url)
#df_prof = pd.read_csv('Computer Science Professors from 55 Top US Schools (Brown University HCI project) - Data.csv')
df_prof.head()
```

To do visualization, we need to compute the following
1. A matrix where matrix[u1][u2] represent the number of people get their doctor's degree from u1 and join u2 as a professor
2. The number of doctors each university produce that joins a top 50 cs university.
3. Number of faculty for each sub-field


```python
universities = df_prof.University.value_counts().keys().tolist()
universitySet = set(universities)
def reformatSchool(school_name):
  return '-'.join(school_name.split('-')[:-1]).rstrip()
```

```python
data = list(df_prof.T.to_dict().values())
```

```python
phdToProf = {}
for professor in data:
  # print(professor['Doctorate'])
  if type(professor['Doctorate']) is str:
    doctor_school = reformatSchool(professor['Doctorate'])
    join_school = professor['University']

    if doctor_school in universitySet:
      if doctor_school not in phdToProf:
        phdToProf[doctor_school] = {}
      if join_school not in phdToProf[doctor_school]:
        phdToProf[doctor_school][join_school] = 0
      phdToProf[doctor_school][join_school] += 1
```

```python
N = len(universities)
matrix_result = [[0] * N for x in range(N)]
```

```python
for i, doctor_school in enumerate(universities):
  for j, join_school in enumerate(universities):
    if doctor_school in phdToProf:
      if join_school in phdToProf[doctor_school]: 
        matrix_result[i][j] = phdToProf[doctor_school][join_school]
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
features = ['Subfield','Bachelors','Doctorate']
```

```python
temp = df_prof[features]
temp.dropna()
temp.head()
```

```python
X_train = np.array(temp)
```

```python
X_train[0]
```

```python
y_train = df_prof.University.tolist()
```

```python
y_train[:10]
```

```python
all_subfields = list(set(df_prof.Subfield.tolist()))
```

```python
universities_set = set(universities)
```

```python
subfield2index = {}
uni2index = {}
for index, field in enumerate(all_subfields):
    subfield2index[field] = index
for index, uni in enumerate(universities):
    uni2index[uni] = index
```

```python
uni2index
```

```python
new_X_train = []
new_y_train = []
for professor, y in zip(X_train,y_train):
    if type(professor[0]) is str and type(professor[1]) is str and type(professor[2]) is str:
        doctor_school = reformatSchool(professor[2])
        bachelor_school = reformatSchool(professor[1])
        print (doctor_school, bachelor_school)
        if doctor_school in universities_set and bachelor_school in universities_set:
            new_X_train.append([subfield2index[professor[0]],uni2index[bachelor_school],uni2index[doctor_school]])
            new_y_train.append(y)
    
```

```python
len(new_X_train)
```

```python
len(all_subfields)
```

```python
len(universities)
```

```python
len(all_universities)
```

```python
import itertools

fieldind = list(range(0, len(all_subfields)))
uniinx = list(range(0, len(universities)))
list_of_features = [fieldind,uniinx,uniinx]
all_combination = list(itertools.product(*list_of_features))
```

```python
len(all_combination)
```

```python
mnb = MultinomialNB()
mnb.fit(new_X_train, new_y_train)
```

```python
result = {}
for field in range(len(all_subfields)):
    result[field] = {}
    for bachelor in range(len(universities)):
        result[field][bachelor] = {}
        for doc in range(len(universities)):
            result[field][bachelor][doc] = 0
all_result = set()
for x in all_combination:
    tmp = mnb.predict([x])
    result[x[0]][x[1]][x[2]] = tmp[0]
    all_result.add(tmp[0])
    
```

```python
len(all_result)
```

```python
import json
with open('predict.json','w') as wf:
    json.dump(result, wf)
```
